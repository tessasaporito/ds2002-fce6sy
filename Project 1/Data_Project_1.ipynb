```python
#imports

import json
import csv
import pandas as pd
import sqlite3
import pymysql
import mysql.connector
from sqlalchemy import create_engine
```

```python
#convert function

#convert to csv file
def to_csv(df, csv_file):
    df.to_csv(csv_file, index=False)
    print(f"Data successfully saved to CSV file")

#convert to json file
def to_json(df, json_file):
    df.to_json(json_file, orient='records', lines=False, indent=4)
    print(f"Data successfully saved to JSON file")

#convert to sqlite database
def to_sql(df, table_name, db_file):
    connection = sqlite3.connect(db_file)
    df.to_sql(table_name, connection, if_exists='replace', index=False)
    connection.close()
    print(f"Data successfully saved to SQLite")

#convert function
def convert():
    df_name = input("Name of dataframe: ")
    try:
        df = eval(df_name)
        if not isinstance(df, pd.DataFrame):
            print(f"{df_name} is not a dataframe")
            return
    except NameError:
        print(f"{df_name} not found.")
        return
    except Exception as e:
        print(f"An error occurred: {e}")
        return
    try:
        target_type = input("Convert into (csv, json, or sql): ")
    except:
        print("Invalid input")
        return
    if target_type == 'csv':
        output_file = input("Output CSV file name: ")
        to_csv(df, output_file)
    elif target_type == 'json':
        output_file = input("Output JSON file name: ")
        to_json(df, output_file)
    elif target_type == 'sql':
        table_name = input("Table name: ")
        db_file = input("Database name: ")
        to_sql(df, table_name, db_file)
    else:
        print("Conversion failed")
```

```python
#store

def store():
    df_name = input("Name of dataframe: ")
    try:
        df = eval(df_name)
        if not isinstance(df, pd.DataFrame):
            print(f"{df_name} is not a dataframe.")
            return
    except NameError:
        print(f"{df_name} not found.")
        return
    except Exception as e:
        print(f"An error occurred: {e}")
        return
    db_file = input("Database name: ")
    table_name = input("Table name: ")
    try:
        connection = sqlite3.connect(db_file)
        df.to_sql(table_name, connection, if_exists='replace', index=False)
        connection.close()
        print(f"Data successfully saved to SQLite database in table '{table_name}'")
    except Exception as e:
        print(f"Storage failed")
```

JSON file

File name: School_Dropout_Rates_in_Iowa__State-wide_and_by_Subgroup.json

File in my GitHub Repo, sourced from Data.gov

```python
#fetch

#put json into df
dropouts = pd.read_json('/School_Dropout_Rates_in_Iowa.json')

dropouts.head()
```

```python
#initial number of columns
print("Number of columns: " + str(dropouts.shape[1]))

#initial number of records
print("Number of records: " + str(dropouts.shape[0]))
```

```python
#modify

#add a column for high/low/average dropout rate (Low: 0 to <1.5, Average: 1.5 to <2.5, High: above 2.5)
def rate_level(rate):
    if rate < 1.5:
        return 'Low'
    elif rate < 2.5:
        return 'Average'
    else:
        return 'High'

dropouts['Level'] = dropouts['Rate'].apply(rate_level)

#Create a Fiscal Year Column based on the Report Period End Column(since all are Oct 1 - Sept 30)
dropouts['Report Period End'] = pd.to_datetime(dropouts['Report Period End'])
dropouts['Fiscal Year'] = dropouts['Report Period End'].dt.year

#drop Report Period Start and Report Period End columns
dropouts = dropouts.drop(columns=['Report Period Start', 'Report Period End'])

dropouts.head()
```

```python
#post processing number of columns
print("Number of columns: " + str(dropouts.shape[1]))

#post processing of records
print("Number of records: " + str(dropouts.shape[0]))
```

```python
#call convert function

convert()
```

```python
#call store function

store()
```
