```python
import pandas as pd

suicide = pd.read_csv('/content/Death_rates_for_suicide__by_sex__race__Hispanic_origin__and_age__United_States.csv')

overdose = pd.read_csv('/content/Drug_overdose_death_rates__by_drug_type__sex__age__race__and_Hispanic_origin__United_States.csv')
```

matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

#line graph for number of male suicides over time
male_suicide = suicide[suicide['STUB_LABEL'] == 'Male']
male_suicide_age_adjusted = male_suicide[male_suicide['UNIT'] == 'Deaths per 100,000 resident population, age-adjusted']
x=male_suicide_age_adjusted['YEAR']
y=male_suicide_age_adjusted['ESTIMATE']

plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o', linestyle='-', color='b', label='Male')
plt.title('Number of Male Suicides Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Suicides (per 100,000 residents, age adjusted)')
plt.legend()
plt.grid(True)
plt.show()
```

plotly

```python
import plotly.express as px
overdose_male = overdose[overdose['STUB_LABEL'] == 'Male']
overdose_female = overdose[overdose['STUB_LABEL'] == 'Female']
overdose_by_sex = pd.concat([overdose_male, overdose_female])

fig = px.bar(
    overdose_by_sex,
    x = 'YEAR',
    y = 'ESTIMATE',
    color = 'STUB_LABEL'
)
fig.show()
```
