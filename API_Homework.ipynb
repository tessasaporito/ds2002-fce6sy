```python
import requests
import pandas as pd
import json

#1
api_key = "x"
stock = input()

#2
url = "https://yfapi.net/v6/finance/quote"
querystring = {"symbols": stock}
headers = {
    'x-api-key': api_key
}

response = requests.get(url, headers=headers, params=querystring)


if response.status_code == 200:
    print("API request successful!")
else:
    print(f"Failed to retrieve data: {response.status_code}")


stock_data = response.json()

if not stock_data['quoteResponse']['result']:
    print(f"Stock symbol '{stock}' not found!")
else:
    stock_info = stock_data['quoteResponse']['result'][0]
    print(f"Ticker: {stock}:")
    print(f"Full Name: {stock_info.get('longName')}")
    print(f"Current Market Price: ${stock_info.get('regularMarketPrice')}")
    print(f"52 Week High: ${stock_info.get('fiftyTwoWeekHigh')}")
    print(f"52 Week Low: ${stock_info.get('fiftyTwoWeekLow')}")

#top 5 trending
url_trending = "https://yfapi.net/v1/finance/trending/us"

response_trending = requests.get(url_trending, headers=headers)

if response_trending.status_code == 200:

    trending_data = response_trending.json()
    trending_stocks = [stock.get('symbol', 'N/A') for stock in trending_data['finance']['result'][0]['quotes'][:5]]
    print("\nTrending Stocks:")
    for stock in trending_stocks:
        print(stock)
else:
    print(f"Failed to retrieve trending stocks: {response_trending.status_code}")
```

```python
#3
df = pd.DataFrame([{
    'Ticker': stock,
    'Full Name': stock_info.get('longName', 'N/A'),
    'Current Market Price': stock_info.get('regularMarketPrice', 'N/A'),
    '52 Week High': stock_info.get('fiftyTwoWeekHigh', 'N/A'),
    '52 Week Low': stock_info.get('fiftyTwoWeekLow', 'N/A')
    }])

df.to_csv(f"{stock}_stock_data.csv")

df.head()
```
